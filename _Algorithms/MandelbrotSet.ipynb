Step 1: Import Required Libraries
Here, we're importing the necessary libraries: numpy for numerical operations and matplotlib.pyplot for creating plots.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Step 2: Define the Mandelbrot Function
This function takes a complex number c and the maximum number of iterations max_iter. It iterates through the Mandelbrot sequence using the equation z = z * z + c and returns the iteration count at which the sequence exceeds a threshold (magnitude greater than 2.0) or reaches the maximum number of iterations.

```python
def mandelbrot(c, max_iter):
    z = c
    for i in range(max_iter):
        if abs(z) > 2.0:
            return i
        z = z * z + c
    return max_iter
```

Step 3: Create Mandelbrot Set Matrix
This function creates the Mandelbrot set by iterating over a grid of complex numbers defined by the ranges of x and y values. For each complex number c, it calculates the iteration count using the mandelbrot function and stores the result in a matrix.

```python
def create_mandelbrot(width, height, xmin, xmax, ymin, ymax, max_iter):
    x_vals = np.linspace(xmin, xmax, width)
    y_vals = np.linspace(ymin, ymax, height)
    mandelbrot_set = np.zeros((height, width))

    for i in range(height):
        for j in range(width):
            x = x_vals[j]
            y = y_vals[i]
            c = x + y * 1j
            mandelbrot_set[i, j] = mandelbrot(c, max_iter)

    return mandelbrot_set
```

Step 4: Plot the Mandelbrot Set
This function plots the Mandelbrot set using imshow from matplotlib. It sets up the plot with appropriate labels and a colorbar.

```python
def plot_mandelbrot(mandelbrot_set, xmin, xmax, ymin, ymax):
    plt.imshow(mandelbrot_set, extent=(xmin, xmax, ymin, ymax), cmap='hot', aspect='equal')
    plt.colorbar()
    plt.title("Mandelbrot Set")
    plt.xlabel("Real")
    plt.ylabel("Imaginary")
    plt.show()
```

Step 5: Set Parameters and Create and Plot the Mandelbrot Set
Here, you define the parameters for the plot (image dimensions, ranges, and max iterations), create the Mandelbrot set using the create_mandelbrot function, and then plot it using the plot_mandelbrot function.

```python
width = 800
height = 800
xmin, xmax = -2.0, 1.0
ymin, ymax = -1.5, 1.5
max_iter = 100

mandelbrot_set = create_mandelbrot(width, height, xmin, xmax, ymin, ymax, max_iter)
plot_mandelbrot(mandelbrot_set, xmin, xmax, ymin, ymax)
```
